# Imputation method

Imputation Methods Replacing missing values with estimated values. Preserves sample size: Doesn’t reduce data points. Can introduce bias: Estimated values might not be accurate.

```python
#  Mean, Median, and Mode Imputation 
mean_imputation = df['atemp'].fillna(df['atemp'].mean()) 
median_imputation = df['atemp'].fillna(df['atemp'].median()) 
mode_imputation = df['atemp'].fillna(df['atemp'].mode().iloc[0]) 
  
print("\nImputation using Mean:") 
print(mean_imputation) 
  
print("\nImputation using Median:") 
print(median_imputation) 
  
print("\nImputation using Mode:") 
print(mode_imputation)
```

# Removal of Outliers

Using IQR, we can follow the below approach to replace the outliers with a NULL value:

Calculate the first and third quartile (Q1 and Q3).
Further, evaluate the interquartile range, IQR = Q3-Q1.
Estimate the lower bound, the lower bound = Q1 x 1.5
Estimate the upper bound, upper bound = Q3 x 1.5
Replace the data points that lie outside of the lower and the upper bound with a NULL value.

For Windspeed

```python
# IQR
Q1 = np.percentile(df['windspeed'], 25, method='midpoint')
Q3 = np.percentile(df['windspeed'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
for x in ['windspeed']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

For hum

```python
# IQR
Q1 = np.percentile(df['hum'], 25, method='midpoint')
Q3 = np.percentile(df['hum'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
for x in ['hum']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

```python
# IQR
Q1 = np.percentile(df['temp'], 25, method='midpoint')
Q3 = np.percentile(df['temp'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
for x in ['temp']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan

```

hus, we have used numpy.percentile() method to calculate the values of Q1 and Q3. Further, we have replaced the outliers with numpy.nan as the NULL values.
